```python
labels = {
    17:0,
    18:1
}
```

```python
w_h = {}

for img_id in cat_dog_img_list:
    for i in data['images']:
        if i['id'] == img_id:
            w_h[img_id] = {'width': i['width'], 'height': i['height']}
```

```python
def coco_to_yolo(bbox, image_width, image_height):
    x, y, width, height = bbox
    x_center = (x + width / 2) / image_width
    y_center = (y + height / 2) / image_height
    norm_width = width / image_width
    norm_height = height / image_height
    return x_center, y_center, norm_width, norm_height
```

```python
for annotation in cat_dog_annotations:
    with open(os.path.join(label_dir, str(annotation[0])+'.txt'), 'a') as f:
        width = w_h[annotation[0]]['width']
        height = w_h[annotation[0]]['height']
        bbox = annotation[1]
        x_center, y_center, norm_width, norm_height = coco_to_yolo(bbox, width, height)
        class_id = labels[annotation[2]]
        yolo_annotation = f"{class_id} {x_center} {y_center} {norm_width} {norm_height}"
        f.write(yolo_annotation + "\n")
```

```python
imgs =[Image.open(filename) for filename in file_list]
```

```python
results = model(imgs)
```

```python
def draw_bbox(draw, bbox, label, color=(0, 255, 0, 255), confs=None, size=15):
    font = ImageFont.truetype("/usr/share/fonts/truetype/humor-sans/Humor-Sans.ttf", size)
    draw.rectangle(bbox, outline=color, width =3)
    def set_alpha(color, value):
        background = list(color)
        background[3] = value
        return tuple(background)
    background = set_alpha(color, 50)
    draw.rectangle(bbox, outline=color, fill=background, width =3)
    background = set_alpha(color, 150)
    text = f"{label}" + ("" if confs==None else f":{conf:0.4}")
    text_bbox = bbox[0], bbox[1], bbox[0]+len(text)*10, bbox[1]+25
    draw.rectangle(text_bbox, outline=color, fill=background, width =3)
    draw.text((bbox[0]+5, bbox[1]+5), text, (0,0,0), font=font)
```

```python
labels={
    0:'cat',
    1:'dog'
}
```

```python
color = []
n_classes = 2
for _ in range(n_classes):
    c = list(np.random.choice(range(256), size=3)) + [255]
    c = tuple(c)
    color.append(c)
```

```python
for i, result in enumerate(results):
    img = imgs[i]
    width, height = img.size
    draw = ImageDraw.Draw(img, 'RGBA')
    result = result.cpu()
    xyxys = result.boxes.xyxyn # normalized í˜•ì‹ìœ¼ë¡œ xyì¢Œí‘œë¥¼ ë¦¬í„´
    confs = result.boxes.conf # confidence score
    clss = result.boxes.cls # cls

    xyxys = xyxys.numpy()
    clss = map(int, clss.numpy())

    for xyxy, conf, cls in zip(xyxys, confs, clss):
        xyxy = [xyxy[0] * width, xyxy[1] * height, xyxy[2] * width, xyxy[3] * height]
        draw_bbox(draw, bbox=xyxy, label=labels[cls], color=color[cls], confs=confs, size=15)
    display(img)
```
