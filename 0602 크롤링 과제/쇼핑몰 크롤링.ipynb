* mysql에 테이블을 만들고, 카테고리와 파일경로를 저장

```python
from selenium import webdriver
import time
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
from urllib.request import Request, urlopen
import os
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)
```

```python
# 네이버 쇼핑 명품지갑/가방 url 접근
url = 'https://search.shopping.naver.com/search/category/100010851'
driver.get(url)
```

```python
datas = driver.find_elements(By.CLASS_NAME, 'product_item__MDtDF')
```

```python
print(len(datas))
```

```python
# 기본 데이터가 5(광고 데이터 포함)개이고, 무한 스크롤 방식이므로, 스크롤을 내림
for _ in range(5):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(1)
```

```python
# 광고 상품을 제외한 요소들 담기
datas = driver.find_elements(By.CLASS_NAME, 'product_item__MDtDF')
```

```python
print(len(datas))
```

```python
products = [] # 데이터 프레임으로 만들 리스트
current_path = os.getcwd() # 현재 폴더의 주소를 저장

# 가져온 상품 만큼 만복
for data in datas:
    # 대 카테고리 -> 중 카테고리 -> 소 카테고리 형식으로 되어있기 때문에, 소 카테고리인 맨 마지막 문자열 가져옴
    category = data.find_elements(By.CLASS_NAME, "product_category__l4FWz")[-1].text
    # 판매하는 상품명 저장
    title = data.find_element(By.CLASS_NAME, 'product_link__TrAac').text
    
    # 해당 폴더가 없으면 폴더 생성
    if not os.path.exists(category):
        os.mkdir(category)
    
    # img태그에 class가 없으므로, 상위 div로 접근
    thumb_div = data.find_element(By.CLASS_NAME, "product_img_area__cUrko")
    # 이미지 주소를 img_url에 저장
    img_url = thumb_div.find_element(By.TAG_NAME, 'img').get_attribute('src')
    # url 마지막이 이상한 숫자로 끝나므로 ?를 기준으로 .jpg까지 가져옴
    filename = img_url.split('/')[-1].split('?')[0]
    
    # 파일 저장
    image_byte = Request(img_url)
    f = open(f'{category}/{filename}', 'wb')
    f.write(urlopen(image_byte).read())
    f.close()
    
    # 파일 경로를 f'{}' 형식으로 하면 \부분이 오류가 나므로, os.path.join을 이용하여 해당 파일 경로 저장
    file_path = os.path.join(current_path, category, filename)
    # 추가할 딕셔너리
    product = {'category' : category , 'title' : title, 'file_path' : file_path}
    products.append(product)
```

```python
print(len(products))
```

```python
# 데이터 프레임으로 저장
df = pd.DataFrame(products)
```
